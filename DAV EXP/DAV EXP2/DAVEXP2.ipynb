```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv')
```

```python
x = df['Hours'].values.reshape(-1, 1)
y = df['Scores'].values.reshape(-1, 1)
```

```python
model = LinearRegression()     #model creation
```

```python
model.fit(x, y)  # fit the data in model
```

```python
print('Coefficients:', model.coef_)
```

```python
print('Intercept:', model.intercept_)
```

```python
plt.figure(figsize=(10, 5))
sns.regplot(x=x, y=y, scatter_kws={'color':'blue'}, line_kws={'color':'red'})
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.title('Student Scores vs Hours')
plt.show()   # ploting data in regression model
```

```python
predicted_scores = model.predict(np.array([[5], [10], [15]]))
print('Predicted scores:', predicted_scores) # score prediction
```
