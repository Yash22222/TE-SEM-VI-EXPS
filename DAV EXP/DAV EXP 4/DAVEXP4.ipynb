```python
import pandas as pd
import plotly.express as px
```

```python
url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
df = pd.read_csv(url)
```

```python
country = 'India'
df_country = df[df['Country/Region'] == country]
```

```python
df_country = df_country.T[4:]
```

```python
df_country = df_country.reset_index()
df_country.columns = ['Date', 'Confirmed Cases']
```

```python
df_country['Date'] = pd.to_datetime(df_country['Date'])
```

```python
fig = px.line(df_country, x='Date', y='Confirmed Cases', title=f'COVID-19 Confirmed Cases in {country}',
              labels={'Confirmed Cases': 'Confirmed Cases'},
              line_shape='linear', render_mode='svg')
fig.update_layout(xaxis_title='Date', yaxis_title='Confirmed Cases', template='plotly_dark')
```
